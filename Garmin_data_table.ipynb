```python
import pandas as pd
import numpy as np
import datetime as dt
```

```python
pd.set_option('display.max_columns', None)
```

```python
df = pd.read_csv('Garmin_cycling_stats_15_21.csv')
df.head()
```

```python
df.drop(df.columns[[2, 13, 14, 15, 16, 17, 18, 19, 21, 22, 23, 24]], axis=1, inplace=True)
df.head()
```

```python
df.rename(columns={'Moving Time':'Moving_Time'}, inplace=True)
df.head()
```

```python
df['Moving_Time_Minutes'] = df['Moving_Time'].str.split(':').apply(lambda x: int(x[0]) * 60 + int(x[1]))
df.head()
```

```python
df.to_csv('bw_garmin_data.csv', index=False)
```
